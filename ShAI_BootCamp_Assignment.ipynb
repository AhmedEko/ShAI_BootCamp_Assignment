```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('Salaries.csv')
df.head()
```

```python
df.columns
```

```python
# check for columns type and no of records
df.info()
```

```python
# Basic statistics
df.describe().T
```

```python
#check for null values
df.isna().sum()
```

```python
#drop 2 columns note and status they all null
df_drop=df.drop(columns =['Status', 'Notes'])
```

```python
df_drop.isna().sum()
```

```python
# median, mode,for BasePay	OvertimePay	OtherPay	Benefits	TotalPay	TotalPayBenefits, determine the range of salaries 
median=df_drop[['BasePay','OvertimePay','OtherPay','Benefits','TotalPay','TotalPayBenefits']].median()
median
```

```python
mode=df_drop[['BasePay','OvertimePay','OtherPay','Benefits','TotalPay','TotalPayBenefits']].mode()
mode
```

```python
salary_range = df_drop['TotalPayBenefits'].describe()[['min', 'max']]
print("Salary Range:", salary_range)
#negative means that he is in debt
```

### data cleaning

```python
#nulls in BasePay   i choose mean to fill basepay not median mean-> avarage salary while median->  middle value in a dataset 
#so it's more appropriate to get the avarage over the middle value
#while  Benefits i choose to fill them with 0 case drop them would be waste of data 
#OvertimePay,OtherPay will drop them as they so small won't affect the data
claned_df=df_drop
claned_df['BasePay'].fillna(claned_df['BasePay'].mean(),inplace=True)
claned_df['Benefits'].fillna(0,inplace=True)
claned_df.dropna(inplace=True)
```

```python
#check again for nulls
claned_df.isna().sum()
```

### Visualization

```python
#histograms 
#X-axis: Represents the numerical values of the corresponding variable.
#Y-axis: Represents the frequency or count of occurrences of values within each bin.
claned_df[['BasePay','OvertimePay','OtherPay','Benefits','TotalPay','TotalPayBenefits']].hist(bins=15,figsize=(15,10))
```

```python
#use pie charts to represent the proportion of employees in different departments.
#pie chart for JobTitle
#some joptitles are the same (Captital letters and small letters)
claned_df.JobTitle.value_counts().head(10).plot(kind='pie',figsize=(10,10),autopct='%1.1f%%',shadow=True)
```

```python
# Group the data by one or more columns and calculate summary statistics for each group, and compare the average salaries across different groups.
#group by jobtitle and get the mean of TotalPayBenefits
claned_df.groupby('JobTitle')['TotalPayBenefits'].mean().sort_values(ascending=False).head(10).plot(kind='barh',figsize=(10,10))
    
```

```python
#Identify any correlation between salary and another numerical column, and plot a scatter plot to visualize the relationship.
#correlation between TotalPayBenefits and TotalPay
claned_df.plot(kind='scatter',x='TotalPayBenefits',y='TotalPay',figsize=(10,10))
    
```

```python
correlation = claned_df['TotalPayBenefits'].corr(claned_df['TotalPay'])
print(correlation)
#positive cor with value colse to 1 means strong linear relationship
```
